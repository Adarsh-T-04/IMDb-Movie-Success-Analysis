```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["rating"], bins=10, color="red", edgecolor="black")
plt.title("Distribution of Rating")
plt.xlabel("Rating")
plt.ylabel("Frequency of Rating")
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df["rating"],color="green")

plt.title("Distribution of Rating")
plt.xlabel("Rating")
plt.ylabel("Frequency of Rating")
plt.grid(alpha=0.3)
plt.show()
```

```python
nm = df["year"].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.plot(nm.index,nm.values, marker="o", linestyle="--", color="magenta")
plt.title("Number of Movies per Year")
plt.ylabel("Count")
plt.xlabel("Year")
plt.show()
```

```python
nm = df["year"].value_counts().sort_index()
plt.figure(figsize=(10,6))

sns.lineplot(x=nm.index , y= nm.values, color="red", marker="o", linestyle="--",label="movies")

plt.title("Number of Movies per Year")
plt.ylabel("Count")
plt.xlabel("Year")
plt.legend()
plt.show()
```

```python
y = df["genre"].str.split(",").explode("genre")
gen = df.groupby(y)["rating"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(x= gen.index, height=gen.values, color="red", width=0.4)
plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Rating")
plt.show()
```

```python
y = df["genre"].str.split(",").explode("genre")
gen = df.groupby(y)["rating"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(x=gen.index, y= gen.values, palette="pastel")

plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(x=df["rating"],y=df["votes"], color="skyblue")

plt.title("Rating VS Votes")
plt.xlabel("Rating")
plt.ylabel("Votes")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x=df["rating"],y=df["votes"], color="skyblue", marker="o", s=100)

plt.title("Rating VS Votes")
plt.xlabel("Rating")
plt.ylabel("Votes")
plt.show()
```
